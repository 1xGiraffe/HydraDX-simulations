```python
import random
import copy
from matplotlib import pyplot as plt
p = [1000] * 1000
p = [p[i-1] + random.random() for i in range(1, len(p))]
alpha = 0.9
o = [p[0]]
for i in range(1, len(p)):
    o.append(o[i-1] * alpha + p[i] * (1-alpha))

# o = [p[i-1] * alpha + p[i] * (1 - alpha) for i in range(1, len(o))]

min_range = 0
max_range = 200
plt.figure(figsize=(20,5))
plt.plot(range(min_range, max_range), p[min_range: max_range], label='price')
plt.plot(range(min_range, max_range), o[min_range: max_range], label='iterative approach')

alt_o = [(1 - alpha) * sum([p[i] * alpha ** (n - i) / (1-alpha ** (n+1)) for i in range(n+1)]) for n in range(len(p))]
plt.plot(range(min_range, max_range), alt_o[min_range: max_range], label='absolute approach')
plt.legend()

plt.figure(figsize=(20, 5))
plt.yscale('log')
plt.plot(range(len(p)), [abs(alt_o[i] / o[i])-1 for i in range(len(p))])
plt.title('percentage difference between absolute and iterative calculations')
```
