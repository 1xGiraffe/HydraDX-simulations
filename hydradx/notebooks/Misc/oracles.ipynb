```python
import random
import copy
from matplotlib import pyplot as plt
p = [100] * 1000
p = [p[i-1] + random.random() for i in range(1, len(p))]
alpha = 0.1
o = [p[0]]
for i in range(1, len(p)):
    o.append(o[i-1] * (1- alpha) + p[i] * alpha)
    
alpha = .9
alt_o = [p[0]]
for i in range(1, len(p)):
    alt_o.append(alt_o[i-1] * alpha + p[i] * (1-alpha))
# o = [p[i-1] * alpha + p[i] * (1 - alpha) for i in range(1, len(o))]

min_range = 100
max_range = 200
plt.figure(figsize=(20,5))
plt.plot(range(min_range, max_range), p[min_range: max_range])
plt.plot(range(min_range, max_range), o[min_range: max_range])
plt.plot(range(min_range, max_range), alt_o[min_range: max_range])

alt_o = [(1 - alpha) * sum([p[i] * alpha ** (n - i) for i in range(n+1)]) for n in range(len(p))]
plt.plot(range(min_range, max_range), alt_o[min_range: max_range])

plt.figure(figsize=(20, 5))
plt.yscale('log')
plt.plot(range(len(p)), [abs(o[i] / alt_o[i])-1 for i in range(len(p))])
plt.title('difference between absolute and iterative calculations')
```
