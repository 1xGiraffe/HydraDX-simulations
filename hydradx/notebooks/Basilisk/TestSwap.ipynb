```python
import pandas as pd
import random

# Experiments
import sys
sys.path.append('../..')
from model import run
from model.plot_utils import *
from model.amm.basilisk_amm import BasiliskPoolState
from model.amm.agents import agent_dict, TradeStrategies

from model import init_utils
from model import processing

# same seed, same parameters = same simulation result
random.seed(42)

initial_state = BasiliskPoolState(
    tokens={
        'R1': 500000,
        'R2': 1500000
    },
    trade_fee=0.1
)

agent_d = {
    'Trader': agent_dict(
        token_list=initial_state.asset_list,
        r_values={'R1': 1000000, 'R2': 1000000},
        trade_strategy=TradeStrategies.random_swaps(amount={'R1': 333, 'R2': 1000}, randomize_amount=True)
    )
}

state = {'state': {'external': {}, 'amm': initial_state, 'agents': agent_d}}
config_dict = init_utils.get_configuration(timesteps=5000)

pd.options.mode.chained_assignment = None  # default='warn'
pd.options.display.float_format = '{:.2f}'.format

run.config(config_dict, state)
events = run.run()

rdf, agent_df = processing.postprocessing(events)
```
